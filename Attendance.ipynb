```python
import numpy as np
import pandas as pd
```

```python
file1='AAPL.csv'
data=pd.read_csv(file1)
data
```

```python
mat=data.to_numpy()
attendence=[]
for i in range(3201,3279):
    occurrences = np.count_nonzero(mat == i)
    attendence.append((i,occurrences))
attendence
```

```python
import csv
with open('finalJuly.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(attendence)
```
